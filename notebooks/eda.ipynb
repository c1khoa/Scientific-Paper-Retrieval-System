```python
import pandas as pd

# 1. Đọc tệp dữ liệu kho bài báo cục bộ
df_papers = pd.read_parquet('../data/raw/papers_2015_2025.parquet')

# 2. Kiểm tra kích thước (Số dòng, Số cột)
print(f"Tổng số lượng bài báo độc lập trong kho: {df_papers.shape[0]:,}")
print(f"Số lượng trường thông tin (cột): {df_papers.shape[1]}")

# 3. Hiển thị thông tin tổng quan về kiểu dữ liệu và bộ nhớ
print("\n--- Thông tin cấu trúc tệp dữ liệu ---")
df_papers.info()
```

```python
# 1. Đọc tệp dữ liệu truy vấn
df_queries = pd.read_parquet('../data/ground_truth/queries.parquet')
print(f"Tổng số lượng câu truy vấn thu được: {len(df_queries)}")

# 2. Tính độ dài từ (số từ) của từng câu truy vấn
df_queries['query_length'] = df_queries['text'].apply(lambda x: len(x.split()))

# 3. Thống kê phân phối độ dài từ (Xem có khớp với Bảng 2.2 không)
print("\n--- Phân bổ số lượng truy vấn theo độ dài từ ---")
length_counts = df_queries['query_length'].value_counts().sort_index()
for length, count in length_counts.items():
    print(f"Truy vấn dài {length} từ: {count} câu")
```

```python
import pandas as pd

# 1. Sao chép bảng dữ liệu
df_test = df_papers.copy()

# 2. Gọi thẳng hàm giải phẳng vì dữ liệu đã là mảng sẵn
all_cats = df_test['categories'].explode()

# 3. Đếm số lượng xuất hiện của từng chuyên ngành
cat_counts = all_cats.value_counts()

print("--- Thống kê số lượng bài báo theo từng phân loại chuyên ngành ---")
for cat in ['cs.AI', 'cs.CL', 'cs.IR', 'cs.LG']:
    count = cat_counts.get(cat, 0)
    print(f"Chuyên ngành {cat}: {count:,} bài báo")
```
